```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Datei hochladen
uploaded = files.upload()

# berprfe die hochgeladenen Dateien
for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')

# Lade die hochgeladene CSV-Datei in ein Pandas DataFrame
df = pd.read_csv(next(iter(uploaded)))
print(df.head())
print(df.info())
print(df.describe())
print(df.columns)
# 1. Deskriptive Statistik
desc_stats = df.describe()
print(desc_stats)

# 2. Korrelationen
# Nur numerische Spalten auswhlen
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)

# Visualisierung der Korrelationen
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 3. Visualisierungen

# Preisverteilung
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Umsatz vs. Anzahl der verkauften Produkte
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Number of products sold', y='Revenue generated', hue='Product type')
plt.title('Revenue vs. Number of Products Sold')
plt.xlabel('Number of Products Sold')
plt.ylabel('Revenue Generated')
plt.show()

# Defektrate nach Produktart
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Product type', y='Defect rates')
plt.title('Defect Rates by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Defect Rates')
plt.show()

# 4. Segmentierte Analysen

# Analyse der Verfgbarkeit nach Standort
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Location', y='Availability')
plt.title('Availability by Location')
plt.xlabel('Location')
plt.ylabel('Availability')
plt.show()

# Analyse der Produktionsvolumina nach Produktart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product type', y='Production volumes')
plt.title('Production Volumes by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Production Volumes')
plt.show()

# 5. Kostenanalyse

# Vergleich der Produktions- und Versandkosten
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Manufacturing costs', y='Shipping costs', hue='Product type')
plt.title('Manufacturing Costs vs. Shipping Costs')
plt.xlabel('Manufacturing Costs')
plt.ylabel('Shipping Costs')
plt.show()

# Gesamtkostenanalyse
df['Total Costs'] = df['Manufacturing costs'] + df['Shipping costs']
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product type', y='Total Costs')
plt.title('Total Costs by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Costs')
plt.show()

# Detaillierte Analyse der Defektraten
plt.figure(figsize=(10, 6))
sns.histplot(df['Defect rates'], bins=20, kde=True)
plt.title('Distribution of Defect Rates')
plt.xlabel('Defect Rates')
plt.ylabel('Frequency')
plt.show()

# Korrelation von Defektraten und Herstellungskosten
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Manufacturing costs', y='Defect rates', hue='Product type')
plt.title('Defect Rates vs. Manufacturing Costs')
plt.xlabel('Manufacturing Costs')
plt.ylabel('Defect Rates')
plt.show()

# Zusammenfassende Statistiken fr verschiedene Segmente
grouped = df.groupby('Product type').mean()
print(grouped)

grouped_location = df.groupby('Location').mean()
print(grouped_location)
```
